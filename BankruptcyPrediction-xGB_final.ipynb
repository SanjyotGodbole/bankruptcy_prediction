## Fitting the xGBoost model

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
# from sklearn import cross_validation, metrics   #Additional scklearn functions
from sklearn import metrics   #Additional scklearn functions
from sklearn.grid_search import GridSearchCV   #Perforing grid search
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix


## store and drop ID
financialData_ID = financialData['ID']
financialData.drop('ID',axis=1,inplace=True)
# split data into X and y
Y = financialData['target']
X = financialData.drop(['target'],axis=1)
```

### Create train and test data

```python
# split data into train and test sets
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)  ## train_test_split() is from sklearn
```

### Custom function to make it easy to fit the model and output the unseen data predictions

```python
## taken from https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/
## and modified for my use-case and setup

def modelfit(alg, y_train, X_train, X_test, y_test,useTrainCV=True, predictUnseen=False,\
                         unseenData=None, cv_folds=5, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(X_train.values, label=y_train.values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='auc', early_stopping_rounds=early_stopping_rounds)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    alg.fit(X_train, y_train,eval_metric='auc')
        
    #Predict test set:
    dtest_predictions = alg.predict(X_test)
    dtest_predprob = alg.predict_proba(X_test)[:,1]
        
    #Print model report:
    print("\nModel Report")
    print("Accuracy : {:f}".format(metrics.accuracy_score(y_test.values, dtest_predictions))) # here, metrics() is from sklearn library
    print("AUC Score (Test): {:f}".format(metrics.roc_auc_score(y_test, dtest_predprob)))
    print("\n\n-----CONFUSION MATRIX------")
    # Creates a confusion matrix
    cm = confusion_matrix(y_test, dtest_predictions)   ## confusion_matrix() is from sklearn library

    # Transform to df for easier plotting
    cm_df = pd.DataFrame(cm,index=['NotBankrupt','Bankrupt'],\
                            columns=['NotBankrupt','Bankrupt'])
    print()
    print(cm_df)
    
    if predictUnseen:
        targetUnseen = alg.predict(unseenData)
    
    xgb.plot_importance(alg)
    return alg, targetUnseen
```

```python
#3 read the data to predict
dataToPredict = pd.read_csv('dataToPredict_no_NA_reduced.csv')
# dataToPredict.head()
dataToPredict_ID = dataToPredict.ID
dataToPredict.drop('ID',axis=1,inplace=True)
dataToPredict.head()
```

#### xgBoost has a hyperparameter called "scale_pos_weight", the typical value for which is given in the documentation:
https://xgboost.readthedocs.io/en/latest/parameter.html

```python
sumPosi = len(y_train[y_train == 1])
sumNeg = len(y_train[y_train==0])
scalePosWt = sumNeg/sumPosi
scalePosWt
```

### Define the model and set the hyperparameters

```python
xgb1 = XGBClassifier(
 learning_rate =0.09,
 n_estimators=1000,
 max_depth=7,
 min_child_weight=1,
 gamma=0,
 subsample=0.75,
 colsample_bytree=0.75,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=scalePosWt,
 seed=27)

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6


## Fit the model and get the fitted model and unseen target predictions
fittedModel, targetUnseenPred = modelfit(xgb1, y_train,X_train,X_test,y_test,predictUnseen=True,unseenData=dataToPredict)
```
